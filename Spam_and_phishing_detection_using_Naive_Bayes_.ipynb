2) Importing libraries

```python
import pandas as pd
import numpy as np
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
```

3) installing 'stopwords' to list common words to be ignored during text analysis (is,are,a,the,and,etc..)

```python
nltk.download('stopwords')
```

4) Importing dataset(https://www.kaggle.com/datasets/uciml/sms-spam-collection-dataset?resource=download)

```python
df = pd.read_csv("spam.csv", encoding='latin-1')
```

Dataset

```python
df
```

5) Selecting specific columns

```python
df = df[['label', 'text']]
```

6)Convert labels: spam → 1, ham (legit) → 0

```python
df['label'] = df['label'].map({'spam': 1, 'ham': 0})
```

7)Text Preprocessing function

```python
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'\W', ' ', text)  # Remove special characters
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    text = text.strip()
    return text
```

```python
df['text'] = df['text'].apply(preprocess_text)
```

8)Splitting the dataset

```python
X_train, X_test, y_train, y_test = train_test_split(df['text'], df['label'], test_size=0.2, random_state=42)
```

9)Feature extraction using TF-IDF(Term Frequency-Inverse Document Frequency)

```python
vectorizer = TfidfVectorizer(stop_words=stopwords.words('english'), max_features=5000)
X_train = vectorizer.fit_transform(X_train)
X_test = vectorizer.transform(X_test)
```

10) Training Naive Bayes model

```python
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)
```

11) Predictions

```python
y_pred = nb_model.predict(X_test)
```

12)Evaluate model

```python
accuracy = accuracy_score(y_test, y_pred)
print("Model Accuracy:", accuracy)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```
